# Q1

Implement a function that checks whether a given string is a palindrome, but with the following constraints:
1. Avoid using any built-in string reversal functions (eg., [::-1]). ✅
2. Do not rely on external libraries or modules. ✅
3. Students should create their own logic for checking palindromes. ✅

```python
"""
Simple recursive function to check if string is palindrome
The input also has left and right pointer for tracking character positions
"""
def checkPalindrome(string: str, left: int, right: int) -> bool:
  if(left >= right):
    return True
  elif(string[left] == string[right]):
    return checkPalindrome(string, left+1, right-1)
  return False

"""
This function transforms all strings to lowercase and removes all spaces before
a string is sent for palindrome check
"""
def cleanInput(string: str) -> str:
  return ''.join(string.split()).lower()
```

## Q1: Give your own input

```python
count = int(input("Enter number of inputs you would like to give: "))
words_arr = []

for i in range(0, count):
  words_arr.append(input(f'Enter # {i + 1} input: '))

for i in range(0, len(words_arr)):
  clean_word = cleanInput(words_arr[i])
  print(f'Is {words_arr[i]} a palindrome: ', checkPalindrome(clean_word, 0, len(clean_word) - 1))
```

## Q1: Given input:

```python
words = ["madam", "APAC", "words", "saippuakivikauppias", "malayalam"]

for i in range(0, len(words)):
  clean_word = cleanInput(words[i])
  print(f'Is "{words[i]}" a palindrome: ', checkPalindrome(clean_word, 0, len(clean_word) - 1))
```

# Q2

Write a Python program that calculates the factorial of a given positive integer, but with the restrictions:
1. that you cannot use any built-in factorial functions or ✅
2. libraries or ✅
3. loops (e.g., math.factorial). ✅

```python
"""
Simple recursive implemention of factorial function with edge cases checking (
  negative inputs not allowed)
"""
def getFactorial(number: int) -> int:
  if(number < 0):
    return "Number must be greater than or equal to 0"
  if(number <= 1 and number >= 0):
    return 1
  else:
    return number * getFactorial(number - 1)


"""
This function just loops and prints the factorial by calling the getFactorial
function
"""
def printFactorial(numbers: int) -> int:
  for i in range(0, len(numbers)):
    print(f'Factorial of {numbers[i]} => ', getFactorial(numbers[i]))
```

## Q2: Give your own input

```python
count = int(input("Enter number of inputs you would like to give: "))
numbers = []

for i in range(0, count):
  numbers.append(int(input(f'Enter # {i + 1} input: ')))

printFactorial(numbers)
```
