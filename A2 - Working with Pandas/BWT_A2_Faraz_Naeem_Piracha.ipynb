## Perform data cleaning (remove duplicates, handle missing values).

```python
# remove duplicates based on all country name
wb_df = wb_df.drop_duplicates(subset=['Country Name', 'Year'], keep='first')
```

```python
# dropping columns with less than 6k records as only 50% of data available.
cols = [7, 10, 14]
wb_df = wb_df.drop(wb_df.columns[cols], axis=1)
```

```python
wb_df.info()
```

```python
# filling missing values appropriately according to their information

methods = {
    'Birth rate, crude (per 1,000 people)': 'mean',
    'Death rate, crude (per 1,000 people)': 'mean',
    'GDP (USD)': 'median',
    'GDP per capita (USD)': 'median',
    'Infant mortality rate (per 1,000 live births)': 'interpolate',
    'Life expectancy at birth (years)': 'median',
    'Population density (people per sq. km of land area)': 'median'
}

def deal_missing_val(df, column, method):
    if method == 'mean':
        df[column] = df[column].fillna(df[column].mean())
    elif method == 'median':
        df[column] = df[column].fillna(df[column].median())
    elif method == 'interpolate':
        df[column] = df[column].interpolate(method='linear')


for column, method in methods.items():
    deal_missing_val(wb_df, column, method)
```

```python
print(wb_df.isnull().sum())
```

## Filter rows based on specific conditions.

```python
print('Countries with GDP per capita > 10000')
display(wb_df[wb_df['GDP per capita (USD)'] > 10000])
```

```python
print('Countries with life expectancy at birth > 70 years')
display(wb_df[wb_df['Life expectancy at birth (years)'] > 80])
```

```python
print('Countries that lie in the "Sub-Saharan Africa" region')
display(wb_df[wb_df['Region'] == 'Sub-Saharan Africa'])
```

```python
print('Countries with GDP per capita > 10000 and Life expectancy at birth > 70 years')
display(wb_df[(wb_df['GDP per capita (USD)'] > 10000) & (wb_df['Life expectancy at birth (years)'] > 75)])
```

```python
print('Countries that are in "Europe & Central Asia" region or are High Income')
display(wb_df[(wb_df['Region'] == 'Europe & Central Asia') | (wb_df['IncomeGroup'] == 'High income')])
```

```python
print('Countries whose name contains the string "stan"')
display(wb_df[wb_df['Country Name'].str.contains('stan', na=False)].drop_duplicates(subset=['Country Name']))
```

## Group data by a categorical variable and calculate summary statistics (mean, count, etc.).

```python
grouped_by_region = wb_df.groupby('Region').agg({
    'Birth rate, crude (per 1,000 people)': ['mean', 'count'],
    'Death rate, crude (per 1,000 people)': ['mean', 'count'],
    'GDP (USD)': ['mean', 'count'],
    'GDP per capita (USD)': ['mean', 'count'],
    'Infant mortality rate (per 1,000 live births)': ['mean', 'count'],
    'Life expectancy at birth (years)': ['mean', 'count'],
    'Population density (people per sq. km of land area)': ['mean', 'count'],
}).reset_index()

display(grouped_by_region)
```

```python
grouped_by_income = wb_df.groupby('IncomeGroup').agg({
    'Birth rate, crude (per 1,000 people)': ['mean', 'count'],
    'Death rate, crude (per 1,000 people)': ['mean', 'count'],
    'GDP (USD)': ['mean', 'count'],
    'GDP per capita (USD)': ['mean', 'count'],
    'Infant mortality rate (per 1,000 live births)': ['mean', 'count'],
    'Life expectancy at birth (years)': ['mean', 'count'],
    'Population density (people per sq. km of land area)': ['mean', 'count'],
}).reset_index()

display(grouped_by_income)
```
