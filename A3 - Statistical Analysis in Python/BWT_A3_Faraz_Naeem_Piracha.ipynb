```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
wb_df = pd.read_excel('./../Data/WorldBank.xlsx')
```

```python
wb_df.head()
```

```python
wb_df.info()
```

```python
def calculate_statistics(column):
    mean_val = wb_df[column].mean()
    median_val = wb_df[column].median()
    std_dev = wb_df[column].std()
    print(f"Statistics for {column}:")
    print(f"Mean: {mean_val}")
    print(f"Median: {median_val}")
    print(f"Standard Deviation: {std_dev}")
    print()
    return mean_val, median_val, std_dev
```

```python
wb_df_drop = wb_df.drop(wb_df.iloc[:, 0:5], axis=1)

for series_name in wb_df_drop.columns:
    calculate_statistics(series_name)
```

```python
def box_hist(column):
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.boxplot(y=wb_df[column])
    plt.title(f'Box plot of {column}')
    
    plt.subplot(1, 2, 2)
    sns.histplot(wb_df[column], kde=True)
    plt.title(f'Histogram of {column}')
    
    plt.show()
```

```python
for series_name in wb_df_drop.columns:
    box_hist(series_name)
```

```python
def skew_kurt(column):
    skewness = wb_df[column].skew()
    kurtosis = wb_df[column].kurt()
    print(f"Skewness for {column}: {skewness}")
    print(f"Kurtosis for {column}: {kurtosis}")
```

```python
for series_name in wb_df_drop.columns:
    skew_kurt(series_name)
```

```python
# Conduct a t-test based on the data provided
def perform_t_test(column, population_mean):
    t_stat, p_val = stats.ttest_1samp(wb_df[column].dropna(), population_mean)
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_val}")
    return t_stat, p_val
```

```python
null_hypothesis = "The mean GDP per capita is equal to $10,000"
alternative_hypothesis = "The mean GDP per capita is different from $10,000"
print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")

# Example: Conducting t-test for GDP per capita with population mean of $10,000
t_stat, p_val = perform_t_test('GDP per capita (USD)', 10000)

# Interpret the p-value and conclude
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis.")
    conclusion = alternative_hypothesis
else:
    print("Fail to reject the null hypothesis.")
    conclusion = null_hypothesis

print(f"Conclusion: {conclusion}")
```
